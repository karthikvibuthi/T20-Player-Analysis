```python
batters = batters.sort_values(by='BATTER')
bowlers = bowlers.sort_values(by='BOWLER')
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

class PlayerInfoExtractor:
    def __init__(self):
        pass

    def extract_team(self, player, url):
        soup = BeautifulSoup(requests.get(url).text, "html.parser")
        Team_and_PlayingRole = soup.find('h1', text=player)
        if not Team_and_PlayingRole:
            return f"Player {player} not found on the page."
        Team_and_PlayingRoleParent_div = Team_and_PlayingRole.parent
        divs = Team_and_PlayingRoleParent_div.find_all('div')
        if not divs:
            return "Team information found."
        spans = divs[0].find_all('span')
        team = spans[0].get_text() if spans else 'N/A'
        return team

    def extract_playing_role(self, player, url):
        soup = BeautifulSoup(requests.get(url).text, "html.parser")
        Team_and_PlayingRole = soup.find('h1', text=player)
        if not Team_and_PlayingRole:
            return f"Player {player} not found on the page."
        Team_and_PlayingRoleParent_div = Team_and_PlayingRole.parent
        divs = Team_and_PlayingRoleParent_div.find_all('div')
        if not divs:
            return "Team information not found."
        spans = divs[0].find_all('span')
        playingRole = spans[2].get_text() if len(spans) > 2 else 'N/A'
        return playingRole

    def extract_batting_style(self, player, url):
        soup = BeautifulSoup(requests.get(url).text, "html.parser")
        BattingStyle = soup.find('p', text='Batting Style')
        if not BattingStyle:
            return "Batting Style information not found."
        BattingStyleParent = BattingStyle.parent
        batting_style_span = BattingStyleParent.find('span')
        Batting_style = batting_style_span.get_text() if batting_style_span else 'N/A'
        return Batting_style

    def extract_bowling_style(self, player, url):
        soup = BeautifulSoup(requests.get(url).text, "html.parser")
        BowlingStyle = soup.find('p', text='Bowling Style')
        if not BowlingStyle:
            return "Bowling Style information not found."
        BowlingStyleParent = BowlingStyle.parent
        bowling_style_span = BowlingStyleParent.find('span')
        Bowling_style = bowling_style_span.get_text() if bowling_style_span else 'N/A'
        return Bowling_style

    def get_player_info(self, player, url):
        team = self.extract_team(player, url)
        playingRole = self.extract_playing_role(player, url)
        Batting_style = self.extract_batting_style(player, url)
        Bowling_style = self.extract_bowling_style(player, url)
        
        data = {
            'Player': player,
            'team': [team],
            'playingRole': [playingRole],
            'Batting_style': [Batting_style],
            'Bowling_style': [Bowling_style]
        }
        
        df = pd.DataFrame(data)
        return df
```

```python
import pandas as pd
import datetime
# Initialize an empty DataFrame to store all player info
all_players_info = pd.DataFrame()

# Initialize the PlayerInfoExtractor class
extractor = PlayerInfoExtractor()

# Iterate over the list of players and URLs
for batters in unique_batters.itertuples():
    try:
        # Get player info DataFrame using the extractor instance
        player_info_df = extractor.get_player_info(batters.BATTER, batters.URL)
        
        # Append player info to the main DataFrame
        all_players_info = pd.concat([all_players_info, player_info_df], ignore_index=True)
        
        # Write the DataFrame to a CSV file
        all_players_info.to_csv('players_info.csv', index=False)
        
    except Exception as e:
        print(f"Error processing {batters.BATTER} from {batters.URL}: {e}")
        fails.append(f"Error processing {bowlers.BOWLER} from {bowlers.URL}: {e}")


```
